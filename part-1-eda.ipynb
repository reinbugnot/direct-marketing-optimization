This notebook outlines the initial Exploratory Data Analysis (EDA) I conducted on the raw dataset `DataScientist_CaseStudy_Dataset.xlsx`, which comprises four sheets (sub-datasets): **Soc_Dem** (socio-demographic information), **Products_ActBalance** (product holdings and volumes), **Inflow_Outflow** (financial transactions), and **Sales_Revenue** (sales and revenue data). The first three sheets represent input features, while the last serves as the target. A thorough inspection and preprocessing of the data were performed to ensure model reliability.

**Duplicate Entry Check:** No duplicate records were found across any dataset.

**Null Value Analysis:** I Identified 28 clients present in `Soc_Dem` and `Products_ActBalance` but missing in `Inflow_Outflow`, indicating dormant accounts. I also further verified that all null values were legitimate (i.e., non-existent transactional records instead of missing), allowing me to replace NaN values with `0` where appropriate.


**Data Type Validation:** I checked for inconsistent data types; only `Sex` was found as a string, requiring label encoding, so I applied label encoding to the `Sex` attribute for model compatibility.

**Outlier Detection & Treatment:** I detected heavy right-skew in continuous numerical features. I applied **Winsorization** on the upper extreme values, capping them at the 95th percentile. Also, I Identified clients under age 16 with substantial account activity. While potentially suspicious, I assumed these are valid data from european banks who allow registration of bank accounts to minors, keeping the entries within the dataset for later training.

**Feature Engineering:** I introduced three new features: `dormant` (flag for dormant accounts), `total_balance` (aggregated balance across accounts), and `credit_debit_ratio` (ratio of credit inflow to debit outflow).

**Feature Redundancy & Multicollinearity Analysis:** I identified high correlation between transaction/volume features and current account transaction/volume features, leading to the removal of redundant current account columns (`_CA` suffix).

**Class Imbalance Assessment:** I found class imbalance in the target labels and planned to address it via class-weight adjustments during modeling.

**Data Export & Next Steps:** I processed `X` (features) and `y` (target labels) and exported them to CSV files, serving as inputs for the modeling and optimization pipeline.

#### V. Check for Outliers
Outliers may represent very high net-worth individuals, mistaken data entries, or fraudulent transactions. We can visually detect these outliers via a boxplot or by plotting the distribution via a histogram. Since most of the features are relatively sparse, I'll exclude zero entries from the corresponding plots to appreciate the distribution better.

```python
# Box plot for each numeric feature
# Ignore zeros

plt.figure(figsize=(15, 7))
X_numeric_no_zeros = X_numeric.copy()
for col in X_numeric_no_zeros.columns:
    X_numeric_no_zeros[col] = X_numeric_no_zeros[col].replace(0, np.nan)
sns.boxplot(data=X_numeric_no_zeros)
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xticks(rotation=90)
plt.title("Boxplot of Numeric Features in Log Scale, ignoring zeros")
plt.show()
```

```python
# Facet grid of histograms for all numeric features
# Ignore zeros

num_rows = (num_cols + 3) // 5  # 4 plots per row, rounded up
fig = plt.figure(figsize=(15, 3*num_rows))

for idx, col in enumerate(X_numeric.columns, 1):
    plt.subplot(num_rows, 5, idx)
    sns.histplot(data=X_numeric_no_zeros[X_numeric_no_zeros[col] != 0][col], bins=30, kde=False)
    plt.title(col)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage of outliers in each numeric feature
Q1 = X_numeric_no_zeros.quantile(0.25)
Q3 = X_numeric_no_zeros.quantile(0.75)
IQR = Q3 - Q1

# outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((X_numeric_no_zeros < lower_bound) | (X_numeric_no_zeros > upper_bound)).sum()
outlier_pct = (outliers / len(X_numeric_no_zeros)) * 100
print("Outlier %:")
print(outlier_pct)
```

We can see some substantial number of outliers in the continuous numerical features in the data. To handle this, I will apply **winsorization**, i.e., I cap the values at both the 5th and 95th percentiles. 

However, the distributions above shows a strong right-skew (many small values, few very large values) on account balances and transaction volumes data. Thus, it makes sense to only cap the upper extreme values at the 95th percentile to preserve the natural distribution while removing potential outliers by adjusting their values down to the 95th percentile value. Based on model performance, I'll know if I need to revisit this step again.

```python
# Winsorization (upper half cap) on select cont. numerical data
for col in (ActBal_list + VolumeCred_list + VolumeDeb_list):
    upper_cap = X[col].quantile(0.95)
    X[col] = np.clip(X[col], None, upper_cap)

# Update intermediary variables
X_numeric = X.select_dtypes(include=['number'])
X_numeric_no_zeros = X_numeric.copy()
for col in X_numeric_no_zeros.columns:
    X_numeric_no_zeros[col] = X_numeric_no_zeros[col].replace(0, np.nan)
```

```python
# Plot the Box plot again for each numeric feature
# Ignore zeros
plt.figure(figsize=(15, 7))
X_numeric_no_zeros = X_numeric.copy()
for col in X_numeric_no_zeros.columns:
    X_numeric_no_zeros[col] = X_numeric_no_zeros[col].replace(0, np.nan)
sns.boxplot(data=X_numeric_no_zeros)
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xticks(rotation=90)
plt.title("Boxplot of Numeric Features in Log Scale, ignoring zeros")
plt.show()
```

##### Inspect Revenue Data

In order to estimate the revenue that I generate from each successful sale, I can do one of two things:
1. Train a separate regressor to estimate how much revenue a product can generate based on the customer's profile (i.e. price is personalized per client)
2. Use an aggregate uniform estimate revenue for each product to calculate expected revenue

Option 1 only works if we have enough useful data to train a regressor with. If our sample size is too small, we risk training a model with very high variance, making its predictions unreliable.

```python
y.loc[:,['Sale_MF', 'Revenue_MF']].dropna().head()
```

```python
nonzero_revenue_MF = ((y['Revenue_MF'].notnull()) & (y['Revenue_MF'] > 0)).sum()
nonzero_revenue_CC = ((y['Revenue_CC'].notnull()) & (y['Revenue_CC'] > 0)).sum()
nonzero_revenue_CL = ((y['Revenue_CL'].notnull()) & (y['Revenue_CL'] > 0)).sum()

print(f'{(nonzero_revenue_MF / len(y) * 100):.1f}% of entries in Revenue_MF are nonzero')
print(f'{(nonzero_revenue_CC / len(y) * 100):.1f}% of entries in Revenue_CC are nonzero')
print(f'{(nonzero_revenue_CL / len(y) * 100):.1f}% of entries in Revenue_CL are nonzero')
```

Here we see that the revenue values are sparse in our dataset - only 12-18% of the 969 valid target labels have revenue greater than zero. This makes sense since revenue can only be positive when there was an actual sale (Sale = 1), and we earlier noted that there's heavy target class imbalance favored against the positive class.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot Revenue_MF distribution with winsorization
revenue_mf_nonzero = y['Revenue_MF'][y['Revenue_MF'] > 0]
revenue_mf_winsorized = np.clip(revenue_mf_nonzero, 
                               revenue_mf_nonzero.quantile(0.0),
                               revenue_mf_nonzero.quantile(0.95))
axes[0].hist(revenue_mf_winsorized, bins=30, edgecolor='black')
axes[0].set_title('Distribution of Non-Zero Revenue_MF (Winsorized)')
axes[0].set_xlabel('Revenue_MF')
axes[0].set_ylabel('Frequency')

# Plot Revenue_CC distribution with winsorization
revenue_cc_nonzero = y['Revenue_CC'][y['Revenue_CC'] > 0]
revenue_cc_winsorized = np.clip(revenue_cc_nonzero,
                               revenue_cc_nonzero.quantile(0.0),
                               revenue_cc_nonzero.quantile(0.95))
axes[1].hist(revenue_cc_winsorized, bins=30, edgecolor='black')
axes[1].set_title('Distribution of Non-Zero Revenue_CC (Winsorized)')
axes[1].set_xlabel('Revenue_CC')
axes[1].set_ylabel('Frequency')

# Plot Revenue_CL distribution with winsorization
revenue_cl_nonzero = y['Revenue_CL'][y['Revenue_CL'] > 0]
revenue_cl_winsorized = np.clip(revenue_cl_nonzero,
                               revenue_cl_nonzero.quantile(0.0),
                               revenue_cl_nonzero.quantile(0.95))
axes[2].hist(revenue_cl_winsorized, bins=30, edgecolor='black')
axes[2].set_title('Distribution of Non-Zero Revenue_CL (Winsorized)')
axes[2].set_xlabel('Revenue_CL')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

From the distributions, we can note the following observations:
1. MF distribution is very right-skewed, with almost all values near zero
2. CC distribution is more symmetric but with notable long tail values (and presence of outliers which we clipped)
3. CL distribution is more spread out


Given the dataset characteristics (high sparsity, skewed distributions, and limited revenue data), it will be difficult to train a separate regressor to make revenue predictions. So instead, a **reliable yet simple** approach is to use the *expected value* of the product itself.

The expected value formula (https://www.investopedia.com/terms/e/expected-value.asp) provides a nice solution to this case study:

$E[X] = P(Y) \times E[X|Y]$

where:
- $E[X]$ represents expected revenue
- $P(Y)$ is the sale probability (from our propensity model)
- $E[X|Y]$ is the mean revenue when a sale occurs

This can be rewritten as:

#### Expected revenue for product $j$ = $P_{i,j} \times$ Avg Revenue Per Sale for Product $j$

where:
- $P_{i,j}$ = propensity score for product $j$ (predicted probability)
- Avg Revenue Per Sale is computed from nonzero revenue values in y.csv
